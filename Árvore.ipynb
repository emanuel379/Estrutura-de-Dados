```python
# Criando uma árvore binária com raiz r

def binary_tree(r):
  return [r, [], []] # Onde os [] dentro de [] são seus filhos

# Inserindo ramo à esquerda

def insert_left(root, new_branch):
  # Analisa se a árvore tem esquerda
  t = root.pop(1)

  # Se a esquerda não for vazia
  if len(t) > 1:
    # inserção na posição 1 da raiz
    root.insert(1, [new_branch, t, []])

  # Se t for vazia
  else:
    root.insert(1, [new_branch, [], []])

  return root

# Inserindo ramo à direita

def insert_right(root, new_branch):
  # Analisa se a árvore tem direita
  t = root.pop(2)

  # Se a direita não for vazia
  if len(t) > 1:
    # inserção na posição 1 da raiz
    root.insert(2, [new_branch, [], t])

  # Se t for vazia
  else:
    root.insert(2, [new_branch, [], []])

  return root

# Recuperando o valor da raiz da árvore
def get_root_val(root):
  return root[0]

# Modificando o valor de uma raíz
def set_root_val(root, new_val):
  root[0] = new_val

def get_left_child(root):
  return root[1]

def get_right_child(root):
  return root[2]
```

```python
# Na forma de classe

class Binarytree:
  # Construtor
  def __init__(self, valor):
    self.key = valor
    self.left_left = None
    self.left_right = None

  # Inserindo nó a esquerda
  def insert_left(self, valor):
    # Se nó corrente não tem filho a esquerda, ok
    if self.left_child == None:
      self.left_child = Binarytree(valor)
    # Se tiver filho a esquerda
    else:
      temp = Binarytree(valor)
      temp.left_child = self.left_child
      self.left_child = temp

  # Inserindo a direita
  def insert_right(self, valor):
    # Se nó corrente não tem filho a esquerda, ok
    if self.right_child == None:
      self.right_child = Binarytree(valor)
    # Se tiver filho a esquerda
    else:
      temp = Binarytree(valor)
      temp.right_child = self.right_child
      self.right_child = temp

  # Obtém o filho da direita
  def get_right_child(self):
    return self.right_child

  # Obtém o filho da esquerda
  def get_left_child(self):
    return self.left_child

  def set_root_val(self, valor):
    self.key = valor

  def get_root_val(self, valor):
    return self.key

# Formas de se adicionar\visitar elementos em uma árvore

  # Percorrer uma árvore binária em Preorder
  def preorder(self):
    # Imprime valor da raiz
    print(self.key)
    # Visita subárvore a esquerda
    if self.left_child:
      self.left.preorder()
    # Visita subárvore a direita
    if self.right_child:
      self.right.preorder()

  # Percorrer uma árvore binária em Inorder
  # Imprime os valores em ordem crescente( em busca binária)
  def inorder(self):
    # Visita subárvore a esquerda
    if self.left_child:
      self.left.inorder()
    # Imprime valor da raiz
    print(self.key)
    # Visita subárvore a direita
    if self.right_child:
      self.right.inorder()

  # Percorrer uma árvore binária em Postorder
  def postorder(self):
    # Visita subárvore a esquerda
    if self.left_child:
      self.left.postorder()
    # Visita subárvore a direita
    if self.right_child:
      self.right.postorder()
    # Imprime valor da raiz
    print(self.key)
```

```python
# Busca binária

import random

# Armazena um nó em uma árvore de busca
class Treenode:
  def __init__(self, key, val, left = None, right = None, parent = None):
    self.key = key
    self.payload = val
    self.left_child = left
    self.right_child = right
    self.parent = parent

  # Verifica se tem filho a esquerda
  def has_left_child(self):
    # Se for None, é porque não tem
    return self.left_child

   # Verifica se tem filho a direita
  def has_right_child(self):
    # Se for None é porque não tem
    return self.right_child

  # Verifica se é filho a direita
  def is_right_child(self):
    # Tem um nó pai e é filho a direita
    return self.parent and self.parent.right_child == self

  # Verifica se é filho a esquerda
  def is_left_child(self):
    # Tem um nó pai e é filho a esquerda
    return self.parent and self.parent.left_child == self

  # Verifica se é raiz
  def is_root(self):
    # Raiz não pode ter pai
    return not self.parent

  # Verifica se é folha
  def is_leaf(self):
    # Raiz não pode ter filho
    return not (self.left_child or self.right_child)

  # Verifica se nó tem filho
  def has_any_children(self):
    return self.left_child or self.right_child

  # Verifica se tem dois filhos
  def has_both_children(self):
    return self.left_child and self.right_child

  # Atualiza dados do nó
  def repladce_node_data(self, key, value, lc, rc):
    self.key = key # Nova chave
    self.payload = value # Novo valor
    # Novos filhos
    self.left_child = lc
    self.right_child = rc
    if self.has_left_child(): # É pai de seu novo filho a esquerda
      self.left_child.parent = self
    if self.has_right_child(): # É pai de seu novo filho a direita
      self.right_child.parent = self

# Implementa a classe Binarysearchtree
class Binarysearchtree:
  def __init__(self):
    self.root = None # Raiz vazia
    self.size = 0

  # Retorna o número de nós da árvore
  def length(self):
    return self.size

  def put(self, key, val):
    # Verificando se já existe uma raiz
    if self.root:
      # Adiciona elementos a partir da raiz
      self._put(key, val, self.root)

    else:
      # Caso a raiz não exista
      self.root = Treenode(key, val)

    # Incrementando no tamanho
    self.size = self.size + 1

  # Função auxiliar recursiva
  def _put(self, key, val, curent):
    # Se a chave é menor, adiciona a esquerda
    if key < curent.key:
      # Se já tiver filho a esquerda, ativa a recursão
      if curent.has_left_child():
        self._put(key, val, curent.left_child)
      else:
        # Está na posição correta
        curent.left_child = Treenode(key, val, parent = curent)

    else:
      # Chave maior ou igual, vai para a direita
      # Se já tiver filho a direita, ativa a recursão
      if curent.has_right_child():
        self._put(key, val, curent.right_child)
      else:
        # Está na posição correta
        curent.right_child = Treenode(key, val, parent = curent)

  # Acrescentando índices na árvore
  def __setitem__(self, k, v):
    self.put(k, v)

  def get(self, key):
     # Se já tiver raiz
     if self.root:
       # Recursão
       res = self._get(key, self.root)
       # Se retorna elemento diferente de None
       if res:
         return res.payload
       else:
         return None
     else:
      return None

  # Função auxiliar recursiva que busca os elementos de uma árvore
  def _get(self, key, curent):
    # se o nó não existe
    if not curent:
      return None
    # se a chave do elemento for igual ao elemento que se busca, então o elemento foi encontrado
    elif curent.key == key:
      return curent
    # Se a chave for menor que a chave nó
    elif key < curent.key:
      # Buscar na subárvore a esquerda
      return self._get(key, curent.left_child)
    # Se a chave for maior que a chave nó
    elif key > curent.key:
      # Buscar na subárvore a direita
      return self._get(key, curent.right_child)

  # Sobrecarregando o operador em colchetes como 'em'
  def __getitem__(self, key):
    return self.get(key)

  # Utilizando o operador in para as operações( verifica se o elemento está presente ou não)
  def __contains__(self, key):
    if self._get(key, self.root):
      return True
    else:
      return False

  # Percorrer uma árvore binária em Preorder
  def preorder(self, curent_node):
    # Imprime valor da raiz
    print(curent_node.key)
    # Visita subárvore a esquerda
    if curent_node.left_child:
      self.preorder(curent_node.left_child)
    # Visita subárvore a direita
    if curent_node.right_child:
      self.preorder(curent_node.right_child)

  # Percorrer uma árvore binária em Inorder
  # Imprime os valores em ordem crescente( em busca binária)
  def inorder(self, curent_node):
    # Visita subárvore a esquerda
    if curent_node.left_child:
      self.inorder(curent_node.left_child)
    # Imprime valor da raiz
    print(curent_node.key)
    # Visita subárvore a direita
    if curent_node.right_child:
      self.inorder(curent_node.right_child)

  # Percorrer uma árvore binária em Postorder
  def postorder(self, curent_node):
    # Visita subárvore a esquerda
    if curent_node.left_child:
      self.postorder(curent_node.left_child)
    # Visita subárvore a direita
    if curent_node.right_child:
      self.postorder(curent_node.right_child)
    # Imprime valor da raiz
    print(curent_node.key)
```
