```python
# Pilha
# Idéia: last in, first out

def pilha():
  b = []
  for i in range(1001):
    if i > 10 and i % 20 == 0:
      b.append(i)
  print(b)

  input()
  # Desempacotando a pilha

  def desempacota():
    for i in range(1, len(b) + 1):
      print(b[-i])

  desempacota()

pilha()
```

```python
# Pilha com classes

class Stack:

  def __init__(self):
    self.items = []

  # Verificando se a pilha está vazia
  def is_empty(self):
    return self.items == []

  # Adicionando elementos no topo
  def push(self, item):
    self.items.append(item)
    print('Push %s' %item)

  # Removendo elementos do topo
  def pop(self):
    print('Pop')
    return self.items.pop()

  # Retornando o elemento do topo sem removêlo
  def peek(self):
    return self.items[-1]

  # Retornando o tamanho da pilha
  def size(self):
    return len(self.items)

  # Imprimindo a pilha
  def print_stack(self):
    print(self.items)

if __name__ =='__main__':

  S = Stack()
  S.print_stack()
  S.push(1)
  S.push(2)
  S.push(3)
  S.print_stack()
  S.pop()
  S.pop()
  S.print_stack()
  S.push(7)
  S.push(8)
  S.push(9)
  S.print_stack()
  print(S.is_empty())
```

```python
# Fila com classes

class Queue:
  # Criando uma fila vazia
  def __init__(self):
    self.items = []

  # Verificando se a fila está vazia
  def is_empty(self):
    return self.items == []

  # Adicionando elemento no início da fila
  def enqueue(self, item):
    self.items.insert(0, item) # Adicionando o último elemento na posição 0
    print('Enqueue %s' %item)

  # Removendo elementos da fila
  def dequeue(self):
    print('Dequeue')
    return self.items.pop()

  # Retornando o tamanho da fila
  def size(self):
    return len(self.items)

  # Imprimindo a fila
  def print_queue(self):
    print(self.items)

if __name__ =='__main__':

  Q = Queue()
  Q.print_queue()
  Q.enqueue(1)
  Q.enqueue(2)
  Q.enqueue(3)
  Q.print_queue()
  Q.dequeue()
  Q.dequeue()
  Q.print_queue()
  Q.enqueue(7)
  Q.enqueue(8)
  Q.enqueue(9)
  Q.print_queue()
  print(Q.is_empty())
```

```python
# Jogo batata-quente com fila
import numpy as np
# Simulando o jogo batata-quente
def batata_quente(names, max):
  # Cria fila que simula a roda
  fila = Queue()
  print(max)
  # Colocando os n nomes na fila
  for name in names:
    fila.enqueue(name)
  # Iniciando a lógica do jogo
  while fila.size() > 1:
    # Simulando o número máximo de passagens da batata
    for i in range(max):
      # removendo o primeiro elemento e colocando-o na última posição
      fila.enqueue(fila.dequeue())
    # Quem parar no início está com a batata deve ser eliminado da fila
    fila.dequeue()

  # Após n-1 rodadas teremos o vencedor
  vencedor = fila.dequeue()

  return vencedor

if __name__ =='__main__':
  # Chamando a função para simular o jogo
  v = batata_quente(['Alex','Júlia','Pedro', 'Damião', 'Arlete', 'Vinícius', 'Caio'], np.random.randint(5, 15))
  print('O vencedor é %s' %v)
```

```python
# invertendo uma fila usando uma pilha

def inverte_fila(Q):
  # Criando uma pilha vazia
  S = Stack()
  # Enquanto houver elementos na fila
  while not Q.is_empty():
    S.push(Q.dequeue())
  # Enquanto houver elementos na pilha
  while not S.is_empty():
    Q.enqueue(S.pop())

  return Q

if __name__ =='__main__':
  # Criando uma fila vazia
  Q = Queue()
  Q.print_queue()
  Q.enqueue(1)
  Q.enqueue(2)
  Q.enqueue(3)
  Q.enqueue(4)
  Q.enqueue(5)
  # Imprimindo a fila
  Q.print_queue()
  IQ = inverte_fila(Q)
  # Imprime a fila invertida
  IQ.print_queue()
```

```python
# Fila de prioridade
# Onde p é a prioridade e quanto menor o p maior a prioridade
# útil para definir grupos prioritários

# Implementando a classe que será utilizada como elemento da fila
class element:
  def __init__(self, valor, prioridade):
    self.item = valor
    self.prioridade = prioridade

  def __str__(self):
    return str(self.item) + ' ' + str(self.prioridade)

class Priorityqueue:
  # Criando uma fila vazia
  def __init__(self):
    self.items = []

  # Verificando se a fila está vazia
  def is_empty(self):
    return self.items == []

  # Adicionando elemento no início da fila
  def enqueue(self, item, prioridade):
    elemento = element(item, prioridade)
    self.items.insert(0, elemento) # Adicionando o último elemento na posição 0
    print('Enqueue %s' %elemento)

  # Removendo elementos da fila
  def dequeue(self):
    if self.is_empty():
      print('Empty Queue')
    else:
      print('Dequeue')
      posicao = 0
      menor = self.items[posicao].prioridade
      # Encontrando o elemento de maior prioridade
      for i in range(self.size()):
        if self.items[i].prioridade < menor:
          posicao = i
          menor = self.items[i].prioridade
    # Removendo elemento da fila
    removido = self.items.pop(posicao)

    return removido.item

  # Retornando o tamanho da fila
  def size(self):
    return len(self.items)

  # Imprimindo a fila
  def print_queue(self):
    L = []
    for j in self.items:
      L.append(j.item)
    print(L)

if __name__ =='__main__':
  # Criando uma fila vazia
  PQ = Priorityqueue()
  PQ.print_queue()
  PQ.enqueue('a', 2)
  PQ.enqueue('b', 3)
  PQ.enqueue('c', 1)
  PQ.enqueue('d', 4)
  PQ.print_queue()
  PQ.dequeue()
  PQ.print_queue()
```
