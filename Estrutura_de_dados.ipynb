```python
# Heapsort
# Tipo O(n!)

%time

def Permutation(lista, n):
  if n == 1:
    print(lista)
  for i in range(n):
    Permutation(lista, n - 1)
    if n % 2 == 0:
      lista[i], lista[n - 1] = lista[n - 1], lista[i]
    else:
      lista[0], lista[n - 1] = lista[n - 1], lista[0]

lista = [1, 2, 3, 4, 5, 6]
n = len(lista)
Permutation(lista, n)


```

```python
# O(2^n)

%time

def Fibonacci(n):
  if n <= 1:
    return n
  return Fibonacci(n - 1) + Fibonacci(n - 2)

Fibonacci(30)
```

```python
# O(n^2)

%time

def Print_Par(lista):
  for i in lista: # n
    for j in lista: # n
      print(f'Items: {i}, {j}')

lista = [1, 2, 3, 4, 5]
Print_Par(lista)
```

```python
# O(n)

%time

def Print_Item(lista):
  for i in lista: # n
    print(f'Item: {i}')

lista = [1, 2, 3, 4, 5]
Print_Item(lista)
```

```python
def Busca_Linear(lista, value):
  for i in range(len(lista)): # O(n), onde n = len(lista)
    if value == lista[i]:
      return i
  raise ValueError('Valor não encontrado')

lista = [1, 3, 5, 7]
Busca_Linear(lista, 5)
```

```python
# O(n^2)

def exemplo(lista):
  for i in range(len(lista)): # O(n)
    for j in range(len(lista)): # O(n)
      if(i != j and lista[i] == lista[j]):
        return True
      else:
        return False

lista = [1, 3, 5, 7]
exemplo(lista)
```

```python
# o(N)*O(M)

def exemplo_2(lista, lista2):
  for i in range(len(lista)): # O(N)
    for j in range(len(lista2)): # O(M) pois len(lista) != len(lista2)
      if lista[i] == lista[j]:
        return True
      else:
        return False

lista = [1, 3, 5, 7]
lista2 = [1, 3, 3, 5, 6, 7, 9, 10]
exemplo_2(lista, lista2)
```

```python
def buscarMenor(lista):
    menor = lista[0]
    count = 0
    for i in lista: # O(n)
      if i < menor:
         menor = i
         count = 0
      if menor == i:
        count += 1
    print(count)
    return menor

lista = [1, 3, 3, 5, 6, 1, 7, 9, 10, 1]
buscarMenor(lista)
```

```python
# 0(1)

%time

def inverter_lista(lista):
  l = []
  for i in lista:
    l.append(lista[-i])
  print(l)

lista = [1, 2 , 3, 4, 5]
inverter_lista(lista)
```
